```python
import os
import imageio
import numpy as np
from matplotlib.pyplot import imshow
from PIL import Image, ImageFont, ImageDraw

from yolo3_predict import YOLO
```

```python
yolo = YOLO()

input_image_name = "test.jpg"
input_image = Image.open("images/" + input_image_name)
save_image = yolo.detect_image(input_image)
# yolo.close_session()

save_image.save(os.path.join("output", input_image_name))
```

```python
output_image = imageio.imread(os.path.join("output", input_image_name))
imshow(output_image)
```
